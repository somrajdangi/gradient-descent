```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([3,5,6,8,9,10])
y = np.array([31,40,60,78,82,89])

costs,weights,bias = [],[],[]
w,b,lr = 0.1,0.01,0.001

for i in range(500):
  yp=x*w+b
  yp = np.dot(x,w)
  cost = np.sum((y-yp)**2)/len(y)
  costs.append(cost)
  weights.append(w)
  bias.append(b)

  dydw = -2/(len(y))* sum(x*(y-yp))
  dydb = -2/(len(y))* sum(y-yp)

  w -=lr*dydw
  b -=lr*dydb
```

```python
plt.plot(weights,costs)
plt.plot(bias,costs)
plt.legend(['w vs cost','bias vs cost'])
plt.show()
```

```python
x1 = np.array([3,5,10]) #study hour
x2 = np.array([5,1,2])  #sleep hour
y = np.array([75,82,93]) #score
n=float(len(x1))
w1,w2,b,lr = 0,0,0,0.0001
costs,w1s,w2s,bs = [],[],[],[]

for i in range(1000):
  yp = w1*x1+w2*x2+b
  
  error = yp-y
  cost = np.sum((error)**2)/n

  costs.append(cost)
  w1s.append(w1)
  w2s.append(w2)
  bs.append(b)

  dydw1 = -2/n *sum(x1*(y-yp))
  dydw2 = -2/n* sum(x2*(y-yp))
  dydb = -2/n*sum(y-yp)

  w1 -=lr*dydw1
  w2 -=lr*dydw2
  b -=lr*dydb

```

```python
plt.plot(w1s,costs)
plt.plot(w2s,costs)
plt.plot(bs,costs)
plt.legend(['w1 vs cost','w2 vs cost','b vs cost'])
plt.show()
```
